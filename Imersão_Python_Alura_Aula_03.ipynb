# Aula 2 - Importação

```python
import pandas as pd
import plotly.express as px

pd.options.display.float_format = "{:,.2f}".format
```

```python
caminho_arquivo = "/content/drive/MyDrive/2024 - Alura - Imersão Python/Tabela de ações - Pura.xlsx"
df_principal = pd.read_excel(caminho_arquivo, sheet_name="Principal")
df_principal.head(10)
```

```python
df_total_acoes = pd.read_excel(caminho_arquivo, sheet_name="Total_de_acoes")
df_total_acoes = df_total_acoes.rename(columns={"Código":"codigo", "Qtde. Teórica":"qtde_teorica"}).copy()
df_total_acoes.head(10)
```

```python
df_ticker = pd.read_excel(caminho_arquivo, sheet_name="Ticker")
df_ticker = df_ticker.rename(columns={"Ticker":"ticker", "Nome":"nome"})
df_ticker.head(10)
```

```python
df_chatgpt = pd.read_excel(caminho_arquivo, sheet_name="ChatGPT")
df_chatgpt = df_chatgpt.drop(columns=["Nome"])
df_chatgpt = df_chatgpt.rename(columns={"Ticker":"ticker", "Segmento":"segmento", "Idade":"idade"})
df_chatgpt
```

```python
df_principal = df_principal[["Ativo","Data","Último (R$)","Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Ativo":"ativo","Data":"data","Último (R$)":"valor_final","Var. Dia (%)":"var_dia_pct"}).copy()
df_principal
```

```python
df_principal["var_pct"] = df_principal["var_dia_pct"] / 100.0
df_principal["valor_inicial"] = df_principal["valor_final"] / (df_principal["var_pct"] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on="ativo", right_on="codigo", how="left")
df_principal = df_principal.drop(columns=["codigo"])
df_principal
```

```python
df_principal["qtde_teorica"] = df_principal["qtde_teorica"].astype(int)
df_principal["variacao_rs"] = (df_principal["valor_final"] - df_principal["valor_inicial"]) * df_principal["qtde_teorica"]
df_principal
```

```python
df_principal["resultado"] = df_principal["variacao_rs"].apply(lambda valor: "Subiu" if valor > 0 else ("Desceu" if valor < 0 else "Estável"))
df_principal
```

```python
df_principal = df_principal.merge(df_ticker, left_on="ativo", right_on="ticker", how="left")
df_principal = df_principal.drop(columns=["ticker"])
df_principal = df_principal.rename(columns={"nome":"nome_da_empresa"})
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on="ativo", right_on="ticker", how="left")
df_principal = df_principal.drop(columns=["ticker", "segmento", "Nome B3"])
df_principal = df_principal.rename(columns={"Segmento B3":"segmento"})
df_principal
```

```python
df_principal["cat_idade"] = df_principal['idade'].apply(lambda idade: "Menos de 50" if idade < 50 else ("Entre 50 e 100" if idade <= 100 else "Mais que 100"))
df_principal
```

```python
maior = df_principal['variacao_rs'].max()
menor = df_principal['variacao_rs'].min()
media = df_principal['variacao_rs'].mean()
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao_rs'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['variacao_rs'].mean()

print(f"Maior: R$ {maior:,.2f}")
print(f"Menor: R$ {menor:,.2f}")
print(f"Média: R$ {media:,.2f}")
print(f"Média de quem subiu: R$ {media_subiu:,.2f}")
print(f"Média de quem desceu: R$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == "Subiu"]
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby("segmento")["variacao_rs"].mean().reset_index()
df_analise_segmento
```

```python
fig = px.pie(df_analise_segmento, names="segmento", values="variacao_rs")
fig.update_xaxes(tickformat=",.2f", tickprefix="R$ ")
fig.show()
```

```python
df_analise_saldo = df_principal.groupby(["resultado"])["variacao_rs"].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x="variacao_rs", y="resultado", text="resultado", title="Variação Reais por Resultado")
fig.update_xaxes(tickformat=",.2f", tickprefix="R$ ")
fig.show()
```

```python
# prompt: Generate a new dataframe df_cat_idade, grouping df_principal by cat_idade, sum column variacao_rs and count the number of rows in each group

df_cat_idade = df_principal.groupby("cat_idade")["variacao_rs"].agg(["sum", "count"]).reset_index()
df_cat_idade = df_cat_idade.rename(columns={"sum":"variacao_rs", "count":"qtde_empresas"})
df_cat_idade
```

```python
fig = px.bar(df_cat_idade.sort_values("variacao_rs"), x="variacao_rs", y="cat_idade", text="cat_idade", title="Variação R$ por Faixa Etária")
fig.update_xaxes(tickformat=",.2f", tickprefix="R$ ")
fig.show()
```

```python
fig = px.bar(df_cat_idade.sort_values("qtde_empresas"), x="qtde_empresas", y="cat_idade", text="cat_idade", title="Quantidade de Empresas por Faixa Etária")
fig.update_xaxes(tickformat=",.0f")
fig.show()
```
